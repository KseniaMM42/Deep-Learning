```python
# 11.10.C.01

# --------------------------------------------------------------------------------
# a) txt-Datei "sentiment_wortliste.txt" mit zuf√§lligen Bewertungeen wird erstellt
# --------------------------------------------------------------------------------

import random

# Liste mit positiven und negativen W√∂rtern
positive_w√∂rter = [
    "hervorragend", "fantastisch", "super", "genial", "ausgezeichnet",
    "perfekt", "toll", "beeindruckend", "wunderbar", "gro√üartig",
    "liebenswert", "n√ºtzlich", "angenehm", "spitze", "erstklassig",
    "exzellent", "brillant", "au√üergew√∂hnlich", "makellos", "tipitopi",
    "sagenhaft", "fabelhaft", "√ºberragend", "wow", "himmlisch",
    "bombastisch", "ph√§nomenal", "sensationell", "traumhaft", "spektakul√§r",
    "un√ºbertroffen", "glorreich", "meisterhaft", "g√∂ttlich", "grandios"   
]

negative_w√∂rter = [
    "entt√§uschend", "schrecklich", "furchtbar", "grauenhaft", "mangelhaft",
    "nervig", "unbrauchbar", "katastrophal", "mies", "ekelhaft",
    "schwach", "lahm", "entsetzlich", "mittelm√§√üig", "langweilig",
    "miserabel", "f√ºrchterlich", "bedauerlich", "deprimierend", "unangenehm", 
    "j√§mmerlich", "frustrierend", "erb√§rmlich", "schauderhaft", "abscheulich", 
    "unm√∂glich", "ungen√ºgend", "nervt√∂tend", "qu√§lend", "√∂de", 
    "grausig", "unbefriedigend", "elend", "abschreckend", "haarstr√§ubend"  
]

# Mische die W√∂rter, damit die Reihenfolge zuf√§llig ist
random.shuffle(positive_w√∂rter)
random.shuffle(negative_w√∂rter)

# Generiere H√§ufigkeiten f√ºr jedes Wort
wortliste = []

for wort in positive_w√∂rter:
    pos_freq = random.randint(50, 150)  # Zuf√§llige H√§ufigkeit f√ºr positive Bewertungen
    neg_freq = random.randint(0, 30)    # Selten negativ bewertet
    wortliste.append((wort, pos_freq, neg_freq))

for wort in negative_w√∂rter:
    pos_freq = random.randint(0, 30)    # Selten positiv bewertet
    neg_freq = random.randint(50, 150)  # Zuf√§llige H√§ufigkeit f√ºr negative Bewertungen
    wortliste.append((wort, pos_freq, neg_freq))

# Liste mischen, damit positive & negative W√∂rter nicht geordnet sind
random.shuffle(wortliste)

# Datei speichern
try:
    with open("sentiment_wortliste.txt", "w", encoding="utf-8") as file:
        file.write("wort\tpositiv\tnegativ\n")  # Spalten√ºberschrift
        for wort, pos, neg in wortliste:
            file.write(f"{wort}\t{pos}\t{neg}\n")
    print("‚úÖ Datei 'sentiment_wortliste.txt' wurde erfolgreich erstellt!")
except Exception as e:
    print(f'\u274C Fehler beim Speichern der Textdatei: {e}')
    
# --------------------------------------------------------------------------------
# b) One-Hot-Encoding der Bewertungen
# --------------------------------------------------------------------------------

import numpy as np

#       (1) Sentiment-Wortliste laden und Vokabular aufbauen
vocabulary = {}  # W√∂rterbuch f√ºr One-Hot-Encoding

try:
    with open("sentiment_wortliste.txt", "r", encoding="utf-8") as file:
        next(file)  # √úberspringe die Kopfzeile
        for index, line in enumerate(file, start=1): # Start bei 1
            word = line.strip().split("\t")[0]  # Erstes Element ist das Wort
            vocabulary[word] = index - 1     # ID bei 0 beginnen lassen
except FileNotFoundError:
    print(f'\u274C Datei wurde nicht gefunden.')
except Exception as e:
    print(f'\u274C Fehler beim Lesen der Textdatei: {e}')

# Anzahl der W√∂rter im Vokabular
vocab_size = len(vocabulary)

#       (2) Funktion zur One-Hot-Codierung einer Bewertung
def one_hot_encode(text, binary=True):
    """
    Wandelt eine Bewertung in einen One-Hot-Vektor um.
    - `binary=True`:  Markiert ein Wort nur einmal (0 oder 1)
    - `binary=False`: Z√§hlt, wie oft jedes Wort vorkommt
    """
    vector = np.zeros(vocab_size)  # Erstelle einen Null-Vektor
    words = text.lower().split()   # Zerlege den Text in W√∂rter
    
    for word in words:
        if word in vocabulary:  # Falls Wort im Vokabular existiert
            index = vocabulary[word]
            if binary:
                vector[index] = 1  # Bin√§re Codierung (nur einmal markieren)
            else:
                vector[index] += 1  # H√§ufigkeitscodierung (z√§hlen)
    return vector

#       Beispielbewertungen
bewertung1 = "Dieses Produkt ist hervorragend und fantastisch"
bewertung2 = "Schrecklich und entt√§uschend"

#       One-Hot-Encoding ausprobieren
vector1 = one_hot_encode(bewertung1, binary=True)
vector2 = one_hot_encode(bewertung2, binary=True)

print("One-Hot-Encoding f√ºr Bewertung 1:", vector1)
print("One-Hot-Encoding f√ºr Bewertung 2:", vector2)

# --------------------------------------------------------------------------------
# c) Neuronales Netz bauen
# --------------------------------------------------------------------------------

from keras.models import Sequential
from keras.layers import Dense

#    Daten -> Bewertungen generiert: synthetische Daten mit zuf√§lligen Wortkombinationen

#       Funktion die positive und negative Bewertungen mit 3 zuf√§lligen W√∂rter generiert
def generiere_bewertung(dominant="positiv"):
    if dominant == "positiv":
        w√∂rter = random.sample(positive_w√∂rter, k=2) + random.sample(negative_w√∂rter, k=1)
    else:
        w√∂rter = random.sample(negative_w√∂rter, k=2) + random.sample(positive_w√∂rter, k=1)
    random.shuffle(w√∂rter)  # Reihenfolge mischen f√ºr Realismus
    return " ".join(w√∂rter)

#       Beispiel-Daten Bewertungen -> 50 klare positive, 50 klare negative, 50 gemischt-positive, 50 gemischt-negative
data = np.array([one_hot_encode(" ".join(random.sample(positive_w√∂rter, k=3)), binary=True) for _ in range(50)] +
                [one_hot_encode(" ".join(random.sample(negative_w√∂rter, k=3)), binary=True) for _ in range(50)] +
                [one_hot_encode(generiere_bewertung("positiv"), binary=True) for _ in range(50)] +
                [one_hot_encode(generiere_bewertung("negativ"), binary=True) for _ in range(50)])

#       Labels (1 = positiv, 0 = negativ)
labels = np.array([1] * 100 + [0] * 100) 

#   (1) Modell erstellen
model = Sequential()
model.add(Dense(10, input_dim=len(vocabulary), activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # Sigmoid f√ºr bin√§re Klassifikation

#   (2) Modell kompilieren
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

#   (3) Modell trainieren
model.fit(data, labels, epochs=10, batch_size=1)

#   (4) Modell evaluieren
loss, accuracy = model.evaluate(data, labels)
print(f'‚úÖ Modellgenauigkeit: {accuracy:.2f}')

# --------------------------------------------------------------------------------
# d) Modellbewertung & Metriken
# --------------------------------------------------------------------------------

from sklearn.metrics import classification_report, precision_score, recall_score, f1_score, confusion_matrix

#       Vorhersagen treffen (0 oder 1)
predictions = (model.predict(data) > 0.5).astype(int)
print(f'Classification report:\n{classification_report(labels, predictions)}')

precision = precision_score(labels, predictions)
recall = recall_score(labels, predictions)
f1 = f1_score(labels, predictions)

print(f"üéØ Pr√§zision: {precision:.2f} - Anteil der als positiv vorhergesagten Bewertungen, die wirklich positiv sind.")
print(f"üîÑ Recall: {recall:.2f} - Anteil der tats√§chlich positiven Bewertungen, die als positiv erkannt wurden.")
print(f"‚öñÔ∏è F1-Score: {f1:.2f} - Harm. Mittelwert von Pr√§zision und Recall, wichtig bei unbalancierten Daten.")

cm = confusion_matrix(labels, predictions)

# Confusion Matrix visualisieren
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Oranges", xticklabels=["Negativ", "Positiv"], yticklabels=["Negativ", "Positiv"])
plt.xlabel("Vorhergesagt")
plt.ylabel("Tats√§chlich")
plt.title("Confusion Matrix")
plt.show()

# --------------------------------------------------------------------------------
# e) Fehleranalyse & Verbesserung
# --------------------------------------------------------------------------------

#   Fehlklassifikationen k√∂nnen durch verschiedene Ursachen entstehen:
#       Modell zu einfach? Mehr Neuronen oder Layers hinzuf√ºgen
#       Datenmenge zu klein? Mehr Trainingsdaten verwenden
#       Vokabular nicht repr√§sentativ? W√∂rter erg√§nzen (z. B. Synonyme)
#       Andere Embeddings? Statt One-Hot z. B. Word2Vec oder GloVe 

# --------------------------------------------------------------------------------
# Grafische Darstellung der Modellgenauigkeit √ºber die Epochen
# --------------------------------------------------------------------------------

# Trainingshistorie auswerten
history = model.fit(data, labels, epochs=10, batch_size=1)

history_dict = history.history
accuracy = history_dict['accuracy']
loss = history_dict['loss']
epochs = range(1, len(accuracy) + 1)

# Genauigkeit und Verlust plotten
plt.figure(figsize=(12, 5))

# Subplot f√ºr die Genauigkeit
plt.subplot(1, 2, 1)
plt.plot(epochs, accuracy, 'bo-', label='Genauigkeit')
plt.xlabel('Epochen')
plt.ylabel('Genauigkeit')
plt.title('Trainingsgenauigkeit √ºber die Epochen')
plt.legend()

# Subplot f√ºr den Verlust
plt.subplot(1, 2, 2)
plt.plot(epochs, loss, 'ro-', label='Verlust')
plt.xlabel('Epochen')
plt.ylabel('Verlust')
plt.title('Trainingsverlust √ºber die Epochen')
plt.legend()

plt.show() 
```
